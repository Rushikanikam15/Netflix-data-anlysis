```python
#q3 year wise data released tv shows and movies
df['date']=pd.to_datetime(df["Release_Date"])
```

```python
df
```

```python
df["Releaseyear"]=pd.DatetimeIndex(df["date"]).year
```

```python
df.head()
```

```python
#q3 year wise data released tv shows and movies
df.groupby("Releaseyear").Category.value_counts()  #shows year wise data                                            
```

```python
#q4 find in which year highest number of tv shows and movies were released.show with bar plot
# horizontal plot
plt.figure(figsize=(9,7))
df["Releaseyear"].value_counts().plot(kind="bar")
plt.title("count movies year ")
plt.show()
```

```python
plt.figure(figsize=(9,7))
df["Releaseyear"].value_counts().plot.pie(autopct="%1.1f%%")
```
